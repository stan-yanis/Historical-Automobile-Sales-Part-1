The dataset used for this visualization assignment contains *historical_automobile_sales* data representing automobile sales and related variables during recession and non-recession period.

<br>1. Date: The date of the observation.
<br>2. Recession: A binary variable indicating recession period; 1 means it was recession, 0 means it was normal.
<br>3. Automobile_Sales: The number of vehicles sold during the period.
<br>4. GDP: The per capita GDP value in USD.
<br>5. Unemployment_Rate: The monthly unemployment rate.
<br>6. Consumer_Confidence: A synthetic index representing consumer confidence, which can impact consumer spending and automobile purchases.
<br>7. Seasonality_Weight: The weight representing the seasonality effect on automobile sales during the period.
<br>8. Price: The average vehicle price during the period.
<br>9. Advertising_Expenditure: The advertising expenditure of the company.
<br>10.Vehicle_Type: The type of vehicles sold; Supperminicar, Smallfamiliycar,                 Mediumfamilycar, Executivecar, Sports.
<br>11.Competition: The measure of competition in the market, such as the number of competitors or market share of major manufacturers.
<br>12.Month: Month of the observation extracted from Date..
<br>13.Year: Year of the observation extracted from Date.
<br>
By examining various factors mentioned above from the dataset, you aim to gain insights into how recessions impacted automobile sales for your company.

### TASK 1.1: Develop a *Line chart* using the functionality of pandas to show how automobile sales fluctuate from year to year


````python
# Since we want to plot automobile sales by year, we need to aggregate the sales data on a yearly basis.
# This involves grouping the data by the 'Year' column and calculating the average of 'Automobile_Sales' for each year.

# Create data for plotting
yearly_sales = df.groupby('Year')['Automobile_Sales'].mean()


# groupby('Year') --> groups the data by 'Year' column
# ['Automobile_sales'].mean() --> calculates the average sales for each year.
yearly_sales.head(5)
````

````python
# Create figure
plt.figure(figsize=(12,6))
yearly_sales.plot(kind='line')
plt.xlabel('Year')
plt.ylabel('Automobile Sales')
plt.title('Average Automobile Sales Per Year')
plt.show()
````

### TASK 1.3: Use the functionality of **Seaborn Library** to create a visualization to compare the sales trend per vehicle type for a recession period with a non-recession period.


````python
# Groupby the values in the 'Recession' column 
# Binary values 1: recession periods and 0: non-recession periods
# After grouping, we select the 'Automobile_Sales' column and apply the .mean() function to each group.
# The reset_index() function is used to convert the result back into a DataFrame.
new_df = df.groupby('Recession')['Automobile_Sales'].mean().reset_index()
new_df
````

````python
# Create Bar chart 'Average Automobile Sales during Recession and Non-Recession'
plt.figure(figsize=(10, 6))
sns.barplot(x='Recession', y='Automobile_Sales', hue='Recession', data=new_df)
plt.xlabel('Non-Recession vs Recession')
plt.ylabel('Automobile Sales')
plt.title('Average Automobile Sales during Recession and Non-Recession')
plt.xticks(ticks=[0, 1], labels=['Non-Recession', 'Recession'])
plt.show()
````

### OPTIONAL : TASK 1.10 Create a map on the hightest sales region/offices of the company during recession period


````python
import requests # The requests library for making HTTP requests

# Creating a function that downloads a file from a specified URL and save it locally with a given filename.
# Download Function
def download(url, filename):
    # Sends an HTTP GET request to the specified URL
    response = requests.get(url)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Open a file in write mode to save the downloaded content
        with open(filename, "wb") as f:
            f.write(response.content) # Write the content of the response to the file
        print(f"{filename} downloaded successfully")
    else:
        # Print an error message if the download fails
        print(f"Failed to download {filename}. Status code: {response.status_code}")

path = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/us-states.json'
filename = "us-states.json"

download(path, filename)


````

#### You found that the datset also contains the location/city for company offices. Now you want to show the recession impact on various offices/city sales by developing a choropleth


````python
# Filter the data for the recession period and specific cities
recession_df.head()
````

````python
sales_by_city = recession_df.groupby('City')['Automobile_Sales'].sum().reset_index()
sales_by_city
````

````python
sales_by_city = recession_df.groupby('City')['Automobile_Sales'].sum()
sales_by_city
````

````python
# Creating a choropleth map using Folium to show the total automobile sales by city during recession periods


# .reset_index(): Converts the grouped data back into a DataFrame with a simple index,
# resulting in a DataFrame (sales_by_city) with two columns: 'City' and 'Automobile_Sales'.

# Calculate the total sales by city
sales_by_city = recession_df.groupby('City')['Automobile_Sales'].sum().reset_index()

# Create a base map centered on the United States
map1 = folium.Map(location=[37.0902, -95.7129], zoom_start=4)

# Create a choropleth layer on the map using Folium 
# to visualize data across geographic regions.
choropleth = folium.Choropleth(
    geo_data = 'us-states.json', # GeoJSON file with state boundaries
    data = sales_by_city,# DataFrame as the data source for the map. 
    columns = ['City', 'Automobile_Sales'],# Indicates which columns from the 'sales_by_city' dataframe to map
    # The first column ('City') is the key that matches regions in the GeoJSON, and the second column ('Automobile_Sales') provides the data to visualize.
    key_on = 'feature.properties.name',# Connects 'City' to the GeoJSON file's 'name' property
    fill_color = 'YlOrRd',
    fill_opacity = 0.7,
    line_opacity = 0.2,
    legend_name = 'Automobile Sales during Recession'
).add_to(map1)

# Add tooltips to the map using a GeoJson layer
folium.GeoJson(
    'us-states.json',
    tooltip=folium.features.GeoJsonTooltip(fields=['name'], aliases=['State:'])
).add_to(map1)

# Display the map (for Jupyter Notebook or IPython environments)
map1

# Display the map
map1
````
